```python
#DATA HANDLING
import pandas as pd
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
data = pd.read_csv('/content/supply_chain_data.csv')
print(data.head())
print(data.describe())
```

```python
X = data[["Order quantities"]]
Y = data["Production volumes"]
Y_ = data["Product type"]
```

```python
#DATA ANALYSIS
plt.scatter(X['Order quantities'], Y, color='b')
plt.xlabel('Order quantities')
plt.ylabel('Production volumes')
plt.show()
```

```python
#LINEAR REGRESSION
mdl = LinearRegression()
mdl.fit(X, Y)
pred = mdl.predict([[6.575]])
print("Predicted value (LR): ",pred[0])
print("Accuracy (LR): ",mdl.score(X[:100], Y[:100])*100)

plt.scatter(X['Order quantities'], Y, color='b')
plt.plot(X['Order quantities'], mdl.predict(X),color='black',linewidth=3)
plt.xlabel('Order quantities')
plt.ylabel('Production volumes')
plt.show()
```

```python
#LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegression
mdl = LogisticRegression()
mdl.fit(X, Y_)
pred = mdl.predict([[6.575]])
print("Predicted value (LGR): ",pred[0])
print("Accuracy (LGR): ",mdl.score(X[:100], Y_[:100])*100)

plt.scatter(X['Order quantities'], Y_, color='b')
plt.plot(X['Order quantities'], mdl.predict(X),color='black',linewidth=3)
plt.xlabel('Order quantities')
plt.ylabel('Production volumes')
plt.show()
```

```python
#SVC
from sklearn.svm import SVC
mdl = SVC(kernel='poly')
mdl.fit(X, Y_)
pred = mdl.predict([[6.575]])
print("Predicted value (SVC): ",pred[0])
#print("Accuracy (SVR): ",mdl.score(X[:100], Y[:100])*100)
plt.scatter(X['Number of products sold'], Y_, color='b')
plt.plot(X['Number of products sold'], mdl.predict(X),color='black',linewidth=3)
plt.xlabel('Number of products sold')
plt.ylabel('Shipping times')
plt.show()
```

```python
#NB
from sklearn.naive_bayes import GaussianNB
mdl = GaussianNB()
mdl.fit(X, Y_)
pred = mdl.predict([[6.575]])
print("Predicted value (NB): ",pred[0])
print("Accuracy (NB): ",mdl.score(X[:100], Y_[:100])*100)

plt.scatter(X['Number of products sold'], Y_, color='b')
plt.plot(X['Number of products sold'], mdl.predict(X),color='black',linewidth=3)
plt.xlabel('Number of products sold')
plt.ylabel('Shipping times)')
plt.show()
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
mdl = KNeighborsClassifier()
mdl.fit(X, Y_)
pred = mdl.predict([[6.575]])
print("Predicted value (KNN): ",pred[0])
print("Accuracy (KNN): ",mdl.score(X[:100], Y_[:100])*100)

plt.scatter(X['Number of products sold'], Y_, color='b')
plt.plot(X['Number of products sold'], mdl.predict(X),color='black',linewidth=3)
plt.xlabel('Number of products sold')
plt.ylabel('Shipping times')
plt.show()
```

```python
#RANDOM FOREST CLASSIFICATION
from sklearn.ensemble import RandomForestClassifier
mdl = RandomForestClassifier(criterion='entropy')
mdl.fit(X, Y_)
pred = mdl.predict([[6.575]])
print("Predicted value (RFC): ",pred[0])
print("Accuracy (RFC): ",mdl.score(X[:100], Y_[:100])*100)

plt.scatter(X['Number of products sold'], Y_, color='b')
plt.plot(X['Number of products sold'], mdl.predict(X),color='black',linewidth=3)
plt.xlabel('Number of products sold')
plt.ylabel('Shipping times')
plt.show()
```

```python
#RANDOM FOREST REGRESSION
from sklearn.ensemble import RandomForestRegressor
mdl = RandomForestRegressor(n_estimators=100,max_depth=6)
mdl.fit(X, Y)
pred = mdl.predict([[6.575]])
print("Predicted value (RFR): ",pred[0])
print("Accuracy (RFR): ",mdl.score(X[:100], Y[:100])*100)

plt.scatter(X['Number of products sold'], Y, color='b')
plt.plot(X['Number of products sold'], mdl.predict(X),color='black',linewidth=3)
plt.xlabel('Number of products sold')
plt.ylabel('Shipping times')
plt.show()
```

```python
#DECISION TREE CLASSIFICATION
from sklearn.tree import DecisionTreeClassifier
mdl = DecisionTreeClassifier(max_leaf_nodes=3, random_state=1)
mdl.fit(X, Y_)
pred = mdl.predict([[6.575]])
print("Predicted value (DTC): ",pred[0])
print("Accuracy (DTC): ",mdl.score(X[:100], Y_[:100])*100)

plt.scatter(X['Number of products sold'], Y_, color='b')
plt.plot(X['Number of products sold'], mdl.predict(X),color='black',linewidth=3)
plt.xlabel('Number of products sold')
plt.ylabel('Shipping times')
plt.show()
```
